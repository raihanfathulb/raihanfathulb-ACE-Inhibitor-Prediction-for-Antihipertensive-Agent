```python
import pandas as pd
import numpy as np
import pickle
import joblib
from sklearn.model_selection import GridSearchCV, train_test_split, cross_val_score
from sklearn.svm import SVR, LinearSVR

from google.colab import drive
drive.mount('/content/drive')
%cd /content/drive/MyDrive/TA
```

```python
# === Load Dataset ===
df_train =  pd.read_csv('Train.csv')

# Memisahkan fitur (X) dan target (y)
X_train = df_train.drop(columns=["Pic50"], axis=1)
y_train = df_train["Pic50"]

# #coba
# df_test = pd.read_csv('Test_scaled.csv')
# X_test = df_test.drop(["Pic50"],axis=1)
# y_test = df_test['Pic50']


# === Load Fitur ===
with open('selected_feature_linear.pkl', 'rb') as file:
    featuredlinear = pickle.load(file)
```

```python
# Filter data
x_trainlinear = X_train[featuredlinear]
x_trainlinear
```

```python
#y_trainlinear = y.ravel()
y_trainlinear = y_train.to_numpy()
```

```python
# === Parameter Tuning dengan GridSearchCV ===
param_grid = {'C': [0.001, 0.01,0.1, 1, 10, 100, 1000],
              'kernel' : ['linear']
              }

#grid_linear = GridSearchCV(LinearSVR(max_iter = 10000), param_grid=param_grid, refit=True, cv=10, scoring='r2', verbose= 3)
grid_linear = GridSearchCV(SVR(), param_grid=param_grid, refit=True, cv=10, scoring='r2', verbose= 3)
grid_linear.fit(x_trainlinear, y_trainlinear)

print('Parameter Terbaik Linear', grid_linear.best_params_)

# === Evaluasi Model ===
# cross_val_score untuk menghitung rata-rata skor R2
score_linear = cross_val_score(grid_linear.best_estimator_, x_trainlinear, y_trainlinear, cv=10, scoring='r2')
score_linear = np.average(score_linear)

print('Skor linear: ',score_linear)
```

```python
joblib.dump(grid_linear.best_estimator_,"grid_linear.best_estimator.pkl")
```

```python
grid_linear.cv_results_
```

```python
# === Hasil GridSearchCV ===
results = grid_linear.cv_results_

for params, mean_test_score in zip(results['params'], results['mean_test_score']):
  print(f"Parameter: {params} -> Skor Mean Test : {mean_test_score:.4f}")
  print()

```

```python
# Parameter default untuk linear
param_default = {
    'C': 1,
    'kernel': 'linear'
    }

model_linear = SVR(**param_default)

"""model_linear = SVR(
    C=param_default['C'],
    kernel=param_default['kernel']
)
"""
#Evaluasi
score_linear_after = cross_val_score(model_linear, x_trainlinear, y_trainlinear, cv=10, scoring='r2')
score_linear_after = np.average(score_linear_after)
print("R2 Score Default Linear:", score_linear_after)
```
